```python
for i in range(len(nifty_it_index.Close)):
    if (i) < len(nifty_it_index.Close):
        a = abs(nifty_it_index.Close.iloc[i] - nifty_it_index.Close.iloc[i - 1]) / nifty_it_index.Close.iloc[i-1]
        if a > (0.02):
            print('True')
```

```python
for i in range(len(nifty_it_index.Volume)):
    if (i) < len(nifty_it_index.Volume):
        a = abs(nifty_it_index.Volume.iloc[i] - nifty_it_index.Volume.iloc[i - 1]) / nifty_it_index.Volume.iloc[i-1]
        if a > (0.1):
            print(i-1, 'True')
```
